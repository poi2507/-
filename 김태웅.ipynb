###  3) 최소 3곳의 터미널을 거치는 택배 배송 과정<br><br>
* 집화터미널 : 택배기사가 택배를 보내려는 개인이나 기업에 방문해 상품을 인수한다.
* 상품이동(1차) : 지화한 상품을 허브터미널로 옮긴다.
* 허브터미널 : 배달지역으로 택배를 분류한다.
* 상품이동(2차) : 분류된 상품을 배달지역의 지점으로 옮긴다.
* 배달터미널 : 배달지역 지점에 도착한 상품을 분류해 받는 사람에게 전달한다.<br><br>
이러한 기존의 택배 수송과정에서 발생하는 문제점은 다수의 터미널을 거치면서 발생하는 물건의 파손과 지역내배송 또한 터미널을 거치면서 시간이 지연된다는 것이다.

```python
우정산업본부=우정산업본부.set_index('지역')

우정산업본부_x=우정산업본부[우정산업본부.columns[1:]]
우정산업본부_y=우정산업본부[['지역내배송']]
우정산업본부_x = (우정산업본부_x - 우정산업본부_x.mean())/우정산업본부_x.std()


우정산업본부_target=우정산업본부_y
우정산업본부_data=우정산업본부_x
```

```python
X_train, X_test, y_train, y_test = train_test_split(우정산업본부_data, 우정산업본부_target ,test_size=0.1)
xgb_model = xgboost.XGBRegressor(n_estimators=100, learning_rate=0.08, gamma=0, subsample=0.75,
                           colsample_bytree=1, max_depth=7)
```

```python
import random
import string
import pandas as pd
import numpy as np
from haversine import haversine
final_data_bigcontest=pd.read_excel('/Users/taewoong/Documents/coding/final_data_bigcontest.xlsx')
final_data_bigcontest=final_data_bigcontest.drop('Unnamed: 0',axis = 1)
junggu = final_data_bigcontest[final_data_bigcontest['시군구명']=='서울특별시 중구']

nowongu = final_data_bigcontest[final_data_bigcontest['시군구명']=='서울특별시 노원구']
nowongu=nowongu.reset_index().drop('index',axis =1)
nowongu
min_len = 1
max_len = 2

random.uniform(final_data_bigcontest[final_data_bigcontest['시군구명']=='서울특별시 중구'].min()['x'],final_data_bigcontest[final_data_bigcontest['시군구명']=='서울특별시 중구'].max()['x'])
def generate_word(length):
    # random 좌표
    jung_list = []
    for i in range(length):
        random_jung_longtitude = random.uniform(final_data_bigcontest[final_data_bigcontest['시군구명']=='서울특별시 중구'].min()['x'],final_data_bigcontest[final_data_bigcontest['시군구명']=='서울특별시 중구'].max()['x'])
        random_jung_latitude  = random.uniform(final_data_bigcontest[final_data_bigcontest['시군구명']=='서울특별시 중구'].min()['y'],final_data_bigcontest[final_data_bigcontest['시군구명']=='서울특별시 중구'].max()['y'])
        jung_list.append((random_jung_latitude,random_jung_longtitude))

    return jung_list


def generate_population(size, min_len, max_len):
    population = []
    for i in range(size):
        # generate words with balanced length
        length = i % (max_len - min_len + 1) + min_len
        population.append(generate_word(length))
    return population

print(generate_word(length=1))
pop = generate_population(size=100, min_len=min_len, max_len=max_len)



def compute_performace(population):
    performance_list = []
    dong_coordinate = {}
    distance = 0
    for i in range(len(junggu)):
        dong_coordinate[junggu.loc[i,'행정동명']] = (junggu.loc[i,'x'], junggu.loc[i,'y'])
        dong_coordinate[junggu.loc[i,'행정동명']]

    for i in range(len(pop)):
        for j in range(len(pop[i])):
            for k in dong_coordinate:
                distance += haversine(pop[i][j],dong_coordinate[k])
        performance_list.append([distance,pop[i]])
        distance = 0


    population_sorted = sorted(performance_list)
    return population_sorted

def select_survivors(population_sorted, best_sample, lucky_few):
    next_generation = []

    for i in range(best_sample):
        next_generation.append(population_sorted[i][1])

    lucky_survivors = random.sample(population_sorted, k=lucky_few)
    for i in range(len(lucky_survivors)):
        next_generation.append(lucky_survivors[i][1])

    # generate new population if next_generation is too small
    while len(next_generation) < best_sample + lucky_few:
        next_generation.append(generate_word(length=random.randint(2,10)))

    random.shuffle(next_generation)
    return next_generation

pop_sorted = compute_performace(pop)
survivors = select_survivors(pop_sorted, best_sample=20, lucky_few=20)




def create_child(individual1, individual2):
    child = []
    min_len_ind = min(len(individual1), len(individual2))
    for i in range(min_len_ind):
        if (int(100 * random.random()) < 50):
            child.append(individual1[i])
        else:
            child.append(individual2[i])
    return child


def create_children(parents, n_child):
    next_population = []
    for i in range(int(len(parents)/2)):
        for j in range(n_child):
            next_population.append(create_child(parents[i], parents[len(parents) - 1 - i]))
    return next_population



children = create_children(survivors, 5)

def mutate_word(word):
    y=random.uniform(final_data_bigcontest[final_data_bigcontest['시군구명']=='서울특별시 중구'].min()['x'],final_data_bigcontest[final_data_bigcontest['시군구명']=='서울특별시 중구'].max()['x'])
    x=random.uniform(final_data_bigcontest[final_data_bigcontest['시군구명']=='서울특별시 중구'].min()['y'],final_data_bigcontest[final_data_bigcontest['시군구명']=='서울특별시 중구'].max()['y'])
    idx = int(random.random() * len(word))
    if (idx == 0):
        word[idx] = (x,y)
    else:
        word[idx] = (x,y)
    return word

def mutate_population(population, chance_of_mutation):
    for i in range(len(population)):
        if random.random() * 100 < chance_of_mutation:
            population[i] = mutate_word(population[i])
    return population

new_generation = mutate_population(population=children, chance_of_mutation=10)

```
